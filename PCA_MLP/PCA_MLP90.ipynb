```python
import pandas as pd
from sklearn import preprocessing
import numpy as np 
from keras.models import Sequential
from keras.layers import Input, Dense
import tensorflow as tf
from sklearn.model_selection import train_test_split
from google.colab import drive
drive.mount('/content/drive')
```

```python

list0=['o',1,95,90,85]
d=[]
x=3
d.append("/content/drive/MyDrive/Colab Notebooks/dataset/PCA/x_train_"+str(list0[x])+".txt")
d.append("/content/drive/MyDrive/Colab Notebooks/dataset/PCA/y_train_"+str(list0[x])+".txt")
d.append("/content/drive/MyDrive/Colab Notebooks/dataset/PCA/x_test_"+str(list0[x])+".txt")
d.append("/content/drive/MyDrive/Colab Notebooks/dataset/PCA/y_test_"+str(list0[x])+".txt")
```

```python
i=0
df0=np.loadtxt(d[i])
df1=np.loadtxt(d[i+1])
df2=np.loadtxt(d[i+2])
df3=np.loadtxt(d[i+3])


# merge files
X = np.concatenate((df0, df2), axis=0)
print(X.shape)

Y = np.concatenate((df1, df3), axis=0)
Y= np.reshape(Y, (Y.size,1))
print(Y.shape)
```

```python
print(df0.shape)
print(df1.shape)
print(df2.shape)
print(df3.shape)
```

```python
input_size=X.shape[1]
model = Sequential()
InputLayer = Input( batch_input_shape=(None, X.shape[1]), name="dense_input", dtype=tf.float32, sparse=False, ragged=False)
model.add(InputLayer)
model.add(Dense(128, batch_input_shape=(None, X.shape[1] ), name="dense", dtype=tf.float32, activation="relu"))
model.add(Dense(39, name="dense_2", dtype=tf.float32, activation="softmax"))
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy' , metrics=['sparse_categorical_accuracy'])
model.summary()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=123)
history = model.fit(X_train, Y_train, epochs=1000,batch_size=512, validation_split=0.05)
```

```python
model.evaluate(X_test, Y_test)
y_pred = model.predict(X_test)
y_true = Y_test
print("y_true's shape: ", y_true.shape)
print("y_pred's shape: ", y_pred.shape)
pred = np.argmax(y_pred, axis = 1)
pred = np.reshape(pred, (len(pred), 1))
from sklearn.metrics import classification_report
print(classification_report(y_true, pred))
```
